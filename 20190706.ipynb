```python
data = iris[pl] 
```

```python
data.plot(kind='kde') 
```

```python
data.plot(kind='hist') 
```

```python
sns.rugplot(data) 
```

```python
x_min = data.min() - 2
x_max = data.max() + 2
```

```python
print(x_min, x_max) 
```

```python
x_axis = np.linspace(x_min, x_max, 100) 
x_axis
```

```python
from scipy.stats import norm 
```

```python
x = np.arange(-2, 2, 0.1)
plt.plot(x, norm(0,1).pdf(x)) 
```

```python
band_width = 0.5
kernel_list = []
for d in data:
    kernel = norm(d, band_width ).pdf(x_axis)
    kernel = kernel / kernel.max()
    kernel_list.append(kernel)

    plt.plot(x_axis, kernel, color='grey', alpha=0.5)
```

```python
sum_of_kde = np.sum(kernel_list, axis=0) 
plt.plot(x_axis, sum_of_kde) 
```

```python
for bw in np.arange(0.25, 1.75, 0.25):
    sns.kdeplot(data, bw=bw, label=bw ) 
```
